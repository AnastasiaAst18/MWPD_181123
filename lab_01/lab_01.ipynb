```python
#Zadanie 1 i 2
def median(data):
    if not data:
        return None
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]

def trimmed_mean(data, p):
    if not data:
        return None
    if not (0 <= p < 0.5):
        raise ValueError("Parametr p musi być z przedziału [0, 0.5)")
    n = len(data)
    sorted_data = sorted(data)
    k = int(n * p)
    trimmed_data = sorted_data[k : n - k]
    if len(trimmed_data) == 0:
        return 0
    return sum(trimmed_data) / len(trimmed_data)

import mystats
import importlib
importlib.reload(mystats)

def main():
    test_values = [10, 2, 30, 4, 50, 6, 70, 8, 90, 1]
    print("TEST ZADANIA 1")
    res_med = mystats.median(test_values)
    print(f"Dane: {test_values}")
    print(f"Obliczona mediana: {res_med}")
    print("TEST ZADANIA 2")

    p_val = 0.2
    res_trim = mystats.trimmed_mean(test_values, p_val)
    print(f"Parametr p: {p_val}")
    print(f"Średnia ucinana: {res_trim}")

if __name__ == 'main':
    main()
```

```python
#Zadanie 3
import pandas as pd

url_iris = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
cols_iris = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_df = pd.read_csv(url_iris, names=cols_iris)
features = iris_df.drop(columns=['class'])
print(f"Średnia:\n{features.mean()}")
print(f"Mediana:\n{features.median()}")
print(f"Odchylenie standardowe:\n{features.std()}")
grouped_iris = iris_df.groupby('class').agg(['mean', 'median', 'std'])
display(grouped_iris)
```

```python
url_red = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
url_white = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
red_wine = pd.read_csv(url_red, sep=';')
red_wine['color'] = 'red'
white_wine = pd.read_csv(url_white, sep=';')
white_wine['color'] = 'white'
wine_df = pd.concat([red_wine, white_wine], ignore_index=True)

#Zadanie 4
wine_features = wine_df.drop(columns=['quality', 'color'])
variances = wine_features.var().sort_values()
print("Cechy posortowane wg wariancji (rosnąco):")
print(variances)
print("Wariancja w podziale na 'quality':")
var_by_quality = wine_df.drop(columns=['color']).groupby('quality').var()
display(var_by_quality)
print("Uśredniona wariancja dla każdej wartości quality:")
print(var_by_quality.mean(axis=1))
print("Odchylenie standardowe w podziale na kolor:")
std_by_color = wine_df.groupby('color').std().drop(columns=['quality'])
display(std_by_color)

#Zadanie 5
print("Przedziały kwartyli (25%, 50%, 75%):")
display(wine_features.quantile([0.25, 0.5, 0.75]))

#Zadanie 6
column_name = 'residual_sugar'
if column_name in wine_df.columns:
    threshold = wine_df[column_name].quantile(0.95)
    high_sugar_wines = wine_df[wine_df[column_name] > threshold]
    print(f"\nWartość progowa (95. percentyl) dla {column_name}: {threshold:.2f}")
    print(f"Liczba znalezionych pozycji: {len(high_sugar_wines)}")
    display(high_sugar_wines.head())
else:
    print(f"BŁĄD: Nie znaleziono kolumny {column_name}. Sprawdź nazwy powyżej!")
```
