```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df_titanic = pd.read_csv(url)
#Zadanie 1.1
missing_pct = df_titanic.isnull().mean() * 100
print("Procent brakujących wartości:\n", missing_pct)
df_titanic['Embarked'] = df_titanic['Embarked'].fillna('U')
#Zadanie 1.2
df_12 = df_titanic.copy()
df_12 = pd.get_dummies(df_12, columns=['Sex', 'Embarked'], drop_first=False)
#Zadanie 1.3
df_vif_1 = df_12.select_dtypes(include=[np.number]).dropna()
def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data["feature"] = df.columns
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(len(df.columns))]
    return vif_data
print("VIF (bez omijania pułapki):")
print(calculate_vif(df_vif_1))
#Zadanie 1.4
df_14 = df_titanic.copy()
df_14 = pd.get_dummies(df_14, columns=['Sex', 'Embarked'], drop_first=True)
df_vif_2 = df_14.select_dtypes(include=[np.number]).dropna()
print("VIF (z ominięciem pułapki - drop_first=True):")
print(calculate_vif(df_vif_2))
```
