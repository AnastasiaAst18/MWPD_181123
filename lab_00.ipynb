```python
#Zadanie 2
import numpy as np
import pandas as pd
base = np.arange(1, 11).reshape(-1, 1)
a = np.arange(1, 11)
macierz = np.power(base, a)
nazwa_k = ['base'] + [f'pow_{i}' for i in range(2, 11)]
df = pd.DataFrame(macierz, columns = nazwa_k)
print(df)
```

```python
#Zadanie 3
print(df.values.base is macierz)
df_shared = pd.DataFrame(macierz, columns = nazwa_k, copy=False)
```

```python
#Zadanie 4
print(df_shared.dtypes)
df_downcasted = df_shared.apply(pd.to_numeric, downcast = 'integer')
print(f"Oryginalny typ: {df_shared['pow_10'].dtype}")
print(f"Typ po downcast: {df_downcasted['pow_10'].dtype}")
```

```python
#Zadanie 5
iris = pd.read_csv('iris.data')
iris
```

```python
#Zadanie 6
iris.columns = [col.replace(' ', '_') for col in iris.columns]
print(iris.head(10))
```

```python
#Zadanie 7
print("Pamięć (standard):", iris.memory_usage().sum())
print("Pamięć (deep):", iris.memory_usage(deep=True).sum())
numeric_cols = iris.select_dtypes(include=['float64']).columns
iris_smaller = iris.copy()
iris_smaller[numeric_cols] = iris_smaller[numeric_cols].astype('float32')
print("Pamięć po optymalizacji (deep):", iris_smaller.memory_usage(deep=True).sum())
```
